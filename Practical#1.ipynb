## 3) Calculating gain and loss
Here I'm calculating possible gain and loss in two cases listed below. I assume that invested money gain value proportionally to the stock and that there are no additional payments. The mathematical notation of formula to calculate gain/loss:  $ stock_{out}/stock_{in}*investment - investment$

```r
investment <- 100
cat("Case 1 - invested in very beginning, sold in very end")
round(eod$Close[nrow(eod)]/eod$Close[1] * investment, 2) - investment
cat("Case 2 - highest investment - invested in day of lowest value, sold in day of highest value")
round(max(eod$Close)/min(eod$Close) * investment, 2) - investment
```

## 4) Calculating return
In this part day-to-day return is calculated and analysis is presented.

```r
# a
# vector is named ret since return is reserved keywoard in R
temp <- eod$Close
ret = (temp[2:length(temp)]-temp[1:length(temp)-1])/temp[1:length(temp)-1]

# Test
# ret[1] == (eod$Close[2]-eod$Close[1])/eod$Close[1]

cat("Total number of return values ", length(ret), "\n")

# b
cat("Minimal return is:", min(ret),"\n")

cat("Maximal return is:", max(ret),"\n")

# c
cat("Number of returns below -4% and abve 4%:", length(ret[ret > 0.04])+length(ret[ret < -0.04]))
```

Main conclusion is that in short, day-to-day period stock value does not change rapidly, except from some unusual situations. Since most of return values are between 2% and -2%, I assume that short term investment in stock does not provide high incomes in general.

## 6) Autocorrelation
In this part I present autocorrelation in the divided dataset. On the plots below we can observe values of autocorrelation for 3 contiguous subsets of the dataset.

```r
# splitting dataset
splitdf <- split(ret, rep(1:3, length.out = length(ret), each = length(ret)%/%3))
eod_1 = splitdf[[1]]
eod_2 = splitdf[[2]]
eod_3 = splitdf[[3]]
cat("Comparing legths of datasets\n")
length(eod_1)
length(eod_2)
length(eod_3)
```

```r
plot_acf <- function(df, name=" "){
    temp = acf(df, lag.max=100, plot=FALSE)
    plot(temp$acf[2:101], 
    type="h", 
    main=name, 
    xlab="Lag",     
    ylab="ACF", 
    ylim=c(-0.1,0.1),
    las=1,
    xaxt="n")
    abline(h=0)
    # Add labels to the x-axis
    x <- c(0:10)*10
    y <- c(0:10)*10
    axis(1, at=x, labels=y)
}
```

```r
par(mfrow=c(2,2)) 
# set 1
plot_acf(eod_1, "Autocorrelation for set 1")
# set 2
plot_acf(eod_2, "Autocorrelation for set 2")
# set 3
plot_acf(eod_3, "Autocorrelation for set 3")
# test autocorrelation for entire return vector in random order
plot_acf(sample(ret), "Returns in random order")
```

### Summary of autocorrelation

Autocorrelation plots shows, that predicting day-to-day return, with previous data is senseless, since even with very short lag, autocorrelation value rarely exceeds 0.1. Event though, autocorrelation values for ordered sets, are significantly higher, then the one in random ordered set, it is still not enough to predict day-to-day return values, relying on past data. Since the autocorrelation is not significant, I can assume that no regular event influencing stock values (especially day-to-day return value) occurs in stock market.

## 7) Computing the Perason correlation coefficient
In this part I load second required datafile, construct new dataframe with stock values in corresponding days from both datasets, and calculating PCC value.

```r
# reading data
nya <- read.csv2(file="./^NYA.txt", header=TRUE, sep=",", dec=".")
nya[[2]] <- as.Date(nya[[2]],format='%d/%m/%Y')
```

```r
# looking for common dates
setDates <- as.Date(intersect(eod$Date.Time, nya$Date.Time), origin = "1970-01-01")
# creating new dataset
df <- data.frame("commonDate" = setDates, 
                 "ftse" = eod$Close[eod$Date.Time %in% setDates], 
                 "nya" = nya$Close[nya$Date.Time %in% setDates])

# computing new return values basig on new sets
temp_nya <- df$nya
temp_ftse <- df$ftse
ret_nya = (temp_nya[2:length(temp_nya)]-temp_nya[1:length(temp_nya)-1])/temp_nya[1:length(temp_nya)-1]
ret_ftse = (temp_ftse[2:length(temp_ftse)]-temp_ftse[1:length(temp_ftse)-1])/temp_ftse[1:length(temp_ftse)-1]

# computing pcc
cat("PCC between day-to-day returns is equal")
cor(ret_nya, ret_ftse)
```

Observed PCC value is relatively high, which is not very surprising. Stock markets in general are bounded with each other, and general for all of them is rather similar.

Next I will check if time zone matters as well. Since we assume, that New York stock influences London stock, we have to compare values from London, to values from the day before in New York (New York Stock closes 5 hours after London stock, so the effect should be visible in London on the next day).

I will perform same computations once again, with shifting values of FTSE returns, to compare it with returns of NYA from the day before.

```r
# getting subset of common dates 
setDates <- as.Date(intersect(eod$Date.Time, nya$Date.Time), origin = "1970-01-01")
# creating dataframe with shifted return values of FTSE
df2 <- data.frame("commonDate" = setDates[1:length(setDates)-1], 
                 "ftse" = (eod$Close[eod$Date.Time %in% setDates])[2:length(setDates)], 
                 "nya" = (nya$Close[nya$Date.Time %in% setDates])[1:length(setDates)-1])

# computing new return values basig on new sets
temp_nya <- df2$nya
temp_ftse <- df2$ftse
ret_nya2 = (temp_nya[2:length(temp_nya)]-temp_nya[1:length(temp_nya)-1])/temp_nya[1:length(temp_nya)-1]
ret_ftse2 = (temp_ftse[2:length(temp_ftse)]-temp_ftse[1:length(temp_ftse)-1])/temp_ftse[1:length(temp_ftse)-1]

# computing pcc
cat("PCC between day-to-day returns with correction by timezone is equal")
cor(ret_nya2, ret_ftse2)
```

Surprisingly, PCC value in second case is significantly lower. Moreover it seems to disagree with hypothesis stated in previous section. There are numerous possible reasons for this. Among them are: relatively long time difference between NYA Closing and FTSE Opening, overlapping working time of NYA and FTSE, influence of Asian markets and so on.

I will try to perform further visual analysis over this data.
