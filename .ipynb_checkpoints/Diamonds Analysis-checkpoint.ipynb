```r
library(ggplot2)
```

```r
smaller <- subset(diamonds,carat<=2.5)
head(smaller)
```

```r
fit <- lm(smaller$price~smaller$carat)
plot(y=smaller$price,x=smaller$carat,pch=".")
abline(fit,col="red")
```

```r
fit_log <- lm(log(smaller$price)~log(smaller$carat))
plot(y=log(smaller$price),x=log(smaller$carat),pch=".")
abline(fit_log,col="red")
```

```r
summary(fit)
summary(fit_log)
```
