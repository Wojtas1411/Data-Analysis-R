```r
# importing libraries
library(ggplot2)
library(class)
library(gmodels)
library(stats)
library(clue)
#library(plotly)
```

```r
data <- read.csv(file="./k-means-exercise.txt", header=TRUE, sep="\t", dec=".", stringsAsFactors=FALSE)
head(data)
```

```r
# remove NA
data <- data[complete.cases(data),]
head(data)
nrow(data)
```

```r
# shuffle dataset and re-label rows
shuffled <- data[sample(nrow(data), replace=FALSE),]
row.names(shuffled) <- c(1:nrow(shuffled))
```

```r
# min max normalization function
normalize <- function(dataset){
    mini = min(dataset)
    maxi = max(dataset)
    dif = maxi-mini
    return((dataset-mini)/dif)
}
#TEST
#normalize(c(1,2,3,4,5))
```

```r
# normalizing data
shuffled_normalized <- shuffled
shuffled_normalized[,-ncol(shuffled)] = as.data.frame(lapply(shuffled[,-ncol(shuffled)], normalize))
# head(shuffled_normalized)
head(shuffled_normalized)
```

```r
features <- c(2,3,4)
split <- 3000
clustered_set <- shuffled_normalized[c(1:split),features]
test_set <- shuffled_normalized[c((split+1):nrow(shuffled_normalized)),features]
nrow(clustered_set)
nrow(test_set)
```

```r
#?kmeans
```

```r
result <- kmeans(clustered_set, 2, iter.max=10)
#result$cluster
sum(result$cluster==1)
sum(result$cluster==2)
```

```r
#?aggregate
```

```r
summary(clustered_set[result$cluster==1,])
summary(clustered_set[result$cluster==2,])
res_clust = factor(result$cluster==1,labels=c('2','1'))
aggregate(clustered_set, by=list(result$cluster),FUN=mean)
```

```r
clustered_config <- data.frame(clustered_set, "label"=res_clust)
head(clustered_config)
```

```r
total <- c()
for(k in c(1:10)){
    temp<-kmeans(clustered_set, k, iter.max=10)
    total<-c(total, temp$tot.withinss)
}
# total
plot(total, type="b")
```

# SVM
Support Vector Machine

```r
library(e1071)
features <- c(2,3,4,15,16)

# split and do stuff 

#mdf <- shuffled[features]
mdf <- shuffled_normalized[features]
split <- 3000
clustered_set2 <- mdf[c(1:split),]
test_set2 <- mdf[c((split+1):nrow(mdf)),]
# good normalization TODO
# minmax <- caret::preprocess(clustered_set2)

# create dataframe
dat <- data.frame(clustered_set2, 
                  "train_label"=as.factor((shuffled_normalized[c(1:split),ncol(shuffled_normalized)]==0)))
head(dat)

#train model
model <- e1071::svm(train_label~.,data=dat, kernel="linear",scale=FALSE)
#model <- e1071::svm(train_label~.,data=dat, kernel="polynomial",degree=2,scale=FALSE)
#model <- e1071::svm(train_label~.,data=dat, kernel="sigmoid",degree=2,scale=FALSE,coef0=0.001)
#model <- e1071::svm(train_label~.,data=dat, kernel="radial",degree=2,scale=FALSE,coef0=0.001)
model
res <- predict(model, test_set2) # get results

# generate test labels
test_labels <- as.factor(shuffled_normalized[c((split+1):nrow(shuffled_normalized)),ncol(shuffled_normalized)]==0)

# compare
mean(res==test_labels)
```
